<h3 style="background-color:#42154d; font-size:250%; text-align:center; font-family:Georgia; color:white"> Data Cleaning</h3> 

```python
df['Diagnosis_Date'] = pd.to_datetime(df['Diagnosis_Date'], format = '%Y-%m-%d')
df['Treatment_Start_Date'] = pd.to_datetime(df['Treatment_Start_Date'], format = '%Y-%m-%d')
```

Converted `Diagnosis_Date` and `Treatment_Start_Date` data type to time format

```python
df.isnull().sum()
```

```python
df = df.drop(['Death_Date', 'Cause_of_Death'], axis = 1)
df['Comorbidities'] = df['Comorbidities'].fillna('None')
```

- Removed `Death_Date` and `Cause_of_Death` columns from dataframe
- Replaced empty values in `Comorbidities` with None

```python
df.duplicated().sum()
```

There are no duplicates in this dataset

<h3 style="background-color:#42154d; font-size:250%; text-align:center; font-family:Georgia; color:white">Exploratory Data Analysis</h3> 

##### **Question 1:** What is the most common medical condition among patients?

```python
plt.figure(figsize = (10,5))
ax = sns.countplot(df, x = 'Cancer_Type', palette = 'viridis')
for container in ax.containers:
   ax.bar_label(container)
plt.xlabel('Cancer Type')
plt.ylabel('Frequency')
plt.title('Medical condition distribution')
plt.show()
```

##### **Insight:** 
Leukemia is the most common cancer type among patients; however, liver, lung, and pancreatic cancers follow closely with only slightly lower occurrences.

##### **Question 2:** What are the outcomes of cancer patients?

```python
group_df = df.groupby('Outcome')['Patient_ID'].count().reset_index().rename(columns = {'Patient_ID':'Frequency'})

plt.figure(figsize = (10,6))
plt.pie(group_df['Frequency'], labels = group_df['Outcome'], autopct = '%1.1f%%')
plt.title('Outcome distribution')
plt.show()
```

##### **Insight:**
Nearly 50% of patients have recovered, while around 40% are still undergoing treatment. Unfortunately, 10% of the patients have died.

##### **Question 3:** How do treatment types relate to survival rates?

```python
plt.figure(figsize = (10,6))
ax = sns.countplot(df, x = 'Treatment_Type', hue = 'Outcome', palette = 'viridis')
for container in ax.containers:
   ax.bar_label(container)
plt.xlabel('Treatment Type')
plt.ylabel('Frequency')
plt.title('Outcome by Treatment Type')
plt.legend(title = 'Outcome', bbox_to_anchor = (1,0.6))
plt.show()
```

##### **Insight:** 
All treatment types are used relatively evenly across patients, with immunotherapy showing a slightly higher usage rate compared to other treatment options.

##### **Question 4:** How do treatment outcomes vary across different medical conditions?

```python
plt.figure(figsize = (10,6))
ax = sns.countplot(df, x = 'Cancer_Type', hue = 'Outcome', palette = 'viridis')
for container in ax.containers:
   ax.bar_label(container)
plt.xlabel('Cancer Type')
plt.ylabel('Frequency')
plt.title('Outcome by Cancer Type')
plt.legend(title = 'Outcome', bbox_to_anchor = (1,0.6))
plt.show()
```

##### **Insight:**
- Leukemia has the highest number of both recovered and under-treatment patients, while pancreatic cancer shows the highest death rate.
- Ovarian cancer stands out with the lowest number of deaths and also has a high recovery rate and under-treatment count. Overall, the distribution of treatment outcomes is fairly consistent across cancer types.



##### **Question 5:** What is the average time gap between diagnosis and treatment initiation?

```python
df['Time_Gap_Before_Treatment'] = df['Treatment_Start_Date'] - df['Diagnosis_Date']
df['Time_Gap_Before_Treatment'] = df['Time_Gap_Before_Treatment'].dt.days
group_df = df.groupby('Time_Gap_Before_Treatment')['Patient_ID'].count().reset_index().rename(columns = {'Patient_ID':'Frequency'})

plt.figure(figsize = (14, 6))
ax = sns.lineplot(group_df, x = 'Time_Gap_Before_Treatment', y = 'Frequency', color = 'purple')
for i, (x, y) in enumerate(zip(group_df['Time_Gap_Before_Treatment'], group_df['Frequency'])):
    if i % 25 == 0: 
        ax.text(x, y + 3, str(y), ha = 'center', va = 'bottom')
plt.xticks(np.arange(0, 200, 10))
plt.xlabel('Time Gap')
plt.title('Time between Diagnosis and Treatment')
plt.show()

print(group_df['Time_Gap_Before_Treatment'].mean())
```

##### **Insight:**
Most of the treatments does not have any fixed days after diagnosis. However, It takes average of 90 days between treatment and diagnosis.

##### **Question 6:** What are the treatment patterns based on cancer stage?

```python
plt.figure(figsize = (10,6))
ax = sns.countplot(df, x = 'Treatment_Type', hue = 'Cancer_Stage', palette = 'viridis')     
for container in ax.containers:
   ax.bar_label(container)            
plt.xlabel('Treatment Type')
plt.ylabel('Frequency')
plt.title('Treatment by Cancer Stage')
plt.legend(title = 'Cancer Stage', bbox_to_anchor = (1,0.6))
plt.show()

```

##### **Question 7:** How do treatment outcomes differ by smoking history?

##### **Question 8:** How are admissions distributed across hospitals?

```python
group_df = df.groupby([df['Hospital'], df['Diagnosis_Date'].dt.year])['Patient_ID'].count().reset_index(name = 'Frequency')

plt.figure(figsize = (12, 6))
ax = sns.barplot(group_df, x = 'Diagnosis_Date', y = 'Frequency', hue = 'Hospital', palette = 'viridis', width = 0.8)
for container in ax.containers:
   ax.bar_label(container, fontsize = 7) 
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.title('Hospital by Admission year')
plt.legend(title = 'Hospital', bbox_to_anchor = (1,0.6))
plt.show()

print('Max Date of Diagnosis:', df.Diagnosis_Date.max())
```

##### **Insight**: 
Patient admissions have been consistent from 2015 to 2023 across all hospitals. Unfortunately, we only have data up to March 2024 to consider for the year 2024.

##### **Question 9:** What are the most common comorbidities among patients?

```python
group_df = df.groupby('Comorbidities')['Patient_ID'].count().reset_index(name = 'Frequency')
plt.pie(group_df['Frequency'], labels = group_df['Comorbidities'], autopct = '%1.1f%%')
plt.title('Comorbidities Distribution')
plt.show()
```

#### **Insight:**
59.5% of patients are suffering from comorbidities. Among them, hypertension is the most common comorbidity, affecting 29.5% of patients. 

##### **Question 10:** Which medical conditions are most common across different ethnic groups?

```python
plt.figure(figsize = (12, 6))
ax = sns.countplot(df, x = 'Ethnicity', hue = 'Cancer_Type', palette = 'viridis', width = 0.9)
for container in ax.containers:
   ax.bar_label(container, fontsize = 7) 
plt.xlabel('Ethnicity')
plt.ylabel('Frequency')
plt.title('Ethnicity by Cancer Type')
plt.legend(title = 'Cancer Type', bbox_to_anchor=(1.0, 0.8), loc='best')
 
plt.show()
```

#### **Insight:**
- South Asian patients have the highest number of leukemia cases compared to other ethnicities and cancer types.
- The highest number of European patients suffer from liver cancer, which is also the highest among all ethnicities.
- The highest number of African patients suffer from colorectal cancer, East Asians from leukemia, and Arabs from breast cancer.
